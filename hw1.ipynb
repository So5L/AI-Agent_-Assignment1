```python
import openai

client = openai.OpenAI()

PROMPT = """
You are a Movie Expert Agent.

You can access a Movie API.

Base URL:
https://nomad-movies.nomadcoders.workers.dev

Available Endpoints:
1) GET /movies
   → Returns a list of popular movies.

2) GET /movies/:id
   → Returns detailed information about a movie by ID.

3) GET /movies/:id/credits
   → Returns cast and crew information for a movie by ID.

You have access to the following functions:

1) get_popular_movies()
   - Calls: GET /movies
   - Use this when the user asks about popular movies.

2) get_movie_details(id: int)
   - Calls: GET /movies/:id
   - Use this when the user asks about a specific movie by ID.

3) get_movie_credits(id: int)
   - Calls: GET /movies/:id/credits
   - Use this when the user asks who starred in a movie.

IMPORTANT RULES:
- You must choose the correct function based on the user's request.
- You must return ONLY the function call in JSON format.
- Do NOT explain.
- Do NOT answer in natural language.
- Output format must be:

{
  "name": "function_name",
  "arguments": {
    "id": 123
  }
}

If the function does not require arguments, use:

{
  "name": "function_name",
  "arguments": {}
}

Now decide which function to call for the following user questions:

1) 지금 인기 있는 영화가 무엇인지 알려줘
2) movie ID 550에 해당하는 영화가 무엇인지 알려줘
3) movie ID 550에 해당하는 영화에 누가 출연하는지 알려줘
"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": PROMPT}],
)

response
```

```python
message = response.choices[0].message.content
message
```
